**Calculadora Básica para Edad**

```python
a = int(input("digitar la edad en años"))
type(a)

if a >= 0 and a <= 20:
  print("Deacuerdo a su edad eres joven")
else:
  print(f"el numero {a} ingresado esta por fuera del rango")

print("/"*4)
```
