```python
#keras
import tensorflow as tf
import numpy as np
import tensorflow.keras as kr


from IPython.core.display import display, HTML

lr = 0.01                  # learning rate
nn = [2, 16, 8, 1]         # n煤mero de neuronas por capa.

# Creamos el objeto que contendr谩 a nuestra red neuronal, como
# secuencia de capas.
model = kr.Sequential()

# A帽adimos la capa 1
l1 = model.add(kr.layers.Dense(nn[1], activation='relu'))

# A帽adimos la capa 2
l2 = model.add(kr.layers.Dense(nn[2], activation='relu'))

# A帽adimos la capa 3
l3 = model.add(kr.layers.Dense(nn[3], activation='sigmoid'))

# Compilamos el modelo, definiendo la funci贸n de coste y el optimizador.
model.compile(loss='mse', optimizer=kr.optimizers.SGD(learning_rate=0.05), metrics=['acc'])

#PIDE QUE SE DEFINA X PARA PODER CORRER EL MODELO
# Define sample data for X and Y
X = np.random.rand(100, 2) # Sample input data with 100 samples and 2 features
Y = np.random.rand(100, 1) # Sample target data with 100 samples and 1 output

# Y entrenamos al modelo. Los callbacks
model.fit(X, Y, epochs=100)
```
